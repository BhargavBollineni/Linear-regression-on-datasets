```python
Y_data=data['sales']
Y_data
```

```python
X_data=data.drop(['Unnamed_0','sales'],axis=1)
X_data.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression(normalize=True)
```

```python
lr.fit(X_data,Y_data)
```

```python
X_data.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(X_data, Y_data, test_size=0.22)
```

```python
Testscore = lr.score(X_data,Y_data)
Testscore
```

```python
New_prediction=lr.predict(X_test)
New_prediction
```

```python
df=pd.DataFrame({'Actual_Data':Y_test,
                'New_predict':New_prediction})
df
```
